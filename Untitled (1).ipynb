4. Which month is the peak?

```python
# Convert date column to datetime
day["Date"] = pd.to_datetime(day["Date"])
print(day["Date"].head())
```

```python
# Create Year-Month column
day["YearMonth"] = day["Date"].dt.to_period("M")

print(day[["Date", "YearMonth"]].head())
```

```python
# Monthly total cases (using New cases)
monthly_cases = (
    day
    .groupby("YearMonth")["New cases"]
    .sum()
    .reset_index()
)

print(monthly_cases)
```

```python
# Find peak month
peak_month = monthly_cases.loc[monthly_cases["New cases"].idxmax()]
```

```python
print("Peak Month:", peak_month["YearMonth"])
print("Total Cases in Peak Month:", peak_month["New cases"])
```

“The peak of the COVID-19 pandemic was observed in the month with the highest number of newly reported cases, indicating the period of maximum spread and healthcare pressure.”

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_cases["YearMonth"].astype(str),
         monthly_cases["New cases"])

plt.xticks(rotation=45)
plt.title("Monthly COVID-19 New Cases (Global)")
plt.xlabel("Month")
plt.ylabel("New Cases")
plt.tight_layout()
plt.show()
```

“The line graph shows a clear rise and fall in monthly COVID-19 cases, helping to visually identify peak periods and wave patterns.”

2.Active Cases Trend

```python
df = pd.read_csv("data/covid_19_clean_complete.csv")

df["Date"] = pd.to_datetime(df["Date"])

df["Active"] = (
    df["Confirmed"]
    - df["Deaths"]
    - df["Recovered"]
)
```

```python
countries = ["India", "US", "Brazil"]
subset = df[df["Country/Region"].isin(countries)]
```

```python
active_trend = subset.pivot_table(
    values="Active",
    index="Date",
    columns="Country/Region"
)
```

```python
print(active_trend.head())
```

```python
plt.figure(figsize=(10,5))

plt.plot(active_trend.index, active_trend["India"], label="India")
plt.plot(active_trend.index, active_trend["US"], label="USA")
plt.plot(active_trend.index, active_trend["Brazil"], label="Brazil")

plt.legend()
plt.title("Active COVID-19 Cases Trend Comparison")
plt.xlabel("Date")
plt.ylabel("Active Cases")
plt.tight_layout()
plt.show()
```

3. Trend Analysis of Active COVID-19 Cases Over Time

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["YearMonth"] = df["Date"].dt.to_period("M")
df["YearMonth"].head()
```

```python
monthly_cases = (
    df.groupby("YearMonth")["Confirmed"]
    .sum()
    .reset_index()
)
```

```python
monthly_cases["Monthly Growth"] = (
    monthly_cases["Confirmed"].diff()
)

print(monthly_cases.head())
```

```python
monthly_cases["Growth %"] = (
    monthly_cases["Confirmed"].pct_change() * 100
)
```

```python
plt.figure(figsize=(7,5))
plt.plot(monthly_cases["YearMonth"].astype(str),
         monthly_cases["Monthly Growth"])

plt.xticks(rotation=45)
plt.title("Monthly COVID-19 Growth Trend")
plt.xlabel("Month")
plt.ylabel("Case Increase")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.bar(monthly_cases["YearMonth"].astype(str),
        monthly_cases["Confirmed"],
        color="green")

plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Cases")
plt.title("Monthly Cases Bar Chart")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
conditions = [
    (df['Date'] < '2020-09-01'),
    (df['Date'] >= '2020-09-01') & (df['Date'] < '2021-03-01'),
    (df['Date'] >= '2021-03-01')
]

# choices = ['Wave 1', 'Wave 2', 'Wave 3']
# df['Wave'] = pd.Series(pd.Categorical(pd.np.select(conditions, choices, default='Other')))
```

```python
import numpy as np
df['Wave'] = np.select(conditions, choices, default='Other')
```

```python
# 3 countries filter
countries = ['India', 'US', 'Brazil']
wave_df = df[df['Country/Region'].isin(countries)]
```

```python
# Aggregate cases per wave
trend = wave_df.groupby(['Wave','Country/Region'])['Confirmed'].max().reset_index()
```

```python
# Seaborn barplot
plt.figure(figsize=(10,6))
sns.barplot(data=trend, x='Wave', y='Confirmed', hue='Country/Region')

plt.title("COVID Wave Comparison")
plt.ylabel("Max Confirmed Cases")
plt.show()
```
